Given a $\mathbb{R}=1$ vector in a 2-dimensional plane and visualize it using the cells below. In the first markdown cell below, write your vector using LaTeX. You should represent it as a system of linear equations. In your methodology discuss the different functions that you have used, the equation of your vectors, and the flowchart for declaring and displaying your vector. Please make sure that your flowchart has only few words and not putting the entire code as it is bad practice. In your results, display and discuss the vector visualization code you made.

$$ A = 5x + 7y  $$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

line = np.array([5, 7])

c = np.arange(0,10)
fig= plt.figure(figsize=(15,15))
plt.xlim(0,20)
plt.ylim(0,20)
plt.quiver(0,0, line[0], line[1], angles='xy', scale_units='xy',scale=1, color='Blue')  #To show the vector direction with an arrow
plt.plot(c*line[0], c*line[1], color='green')                                             #To show the destination of the vector on the graph   
plt.title('Rank 1 Vector')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(color='red')
plt.show()
```

Given a $\mathbb{R}=2$ vector in a 3-dimensional plane and visualize it using the cells below. In the first markdown cell below, write your vector using LaTeX. You should represent it as a system of linear equations. In your methodology discuss the different functions that you have used, the equation of your vectors, and the flowchart for declaring and displaying your vector. Please make sure that your flowchart has only few words and not putting the entire code as it is bad practice. In your results, display and discuss the vector visualization code you made.

$$Vectors = \left\{ \begin{array}\\ x + 7y + 2z \\ 9x + 5y + 5z \end{array} \right.$$

```python
A= np.array([ [1,7,2],
            [9,5,5]])
colors = ['Green','Purple']

fig= plt.figure(figsize=(15,15))
ax = fig.gca(projection='3d')
ax.set_xlim([0,16])
ax.set_ylim([0,16])
ax.set_zlim([0,16])

c = np.arange(0,3)
ax.quiver(0, 0, 0, A[0,0], A[0,1],A[0,2], arrow_length_ratio=.1,color='Blue',label='Vector 1')           #To show the vector direction with an arrow
ax.quiver(0, 0, 0, A[1,0], A[1,1],A[1,2], arrow_length_ratio=.1,color='Red',label='Vector 2')
for i in range(A.shape[0]):
    ax.plot(c*A[i,0],c*A[i,1],c*A[i,2], color=colors[i])                                                 #To show the destination of the vector on the graph
    
plt.title('Rank 2 Vectors')
plt.legend()
```
