```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Task 1

$$ Equations = \left\{ \begin{matrix} 4x + 5y \\ 10x - 5y\\ \end{matrix} \right. $$

$$X=\begin{bmatrix} 4,10 \end{bmatrix}\\ Y=\begin{bmatrix} 5,-5\\ \end{bmatrix}  $$

```python
vectX = np.array([4,10])
vectY = np.array([5,-5])
a = np.arange(-10,10,.1)
b = np.arange(-10,10,.2)

plt.scatter(a*vectX[0],a*vectX[1], label='Vector X',color='red')
plt.scatter(b*vectY[0],b*vectY[1], label='Vector Y',color='Green')

plt.xlim(-10,10)
plt.ylim(-10,10)
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.grid()
plt.legend()
plt.show()
```

# Task 2 

## Span 1 (RANK 2)
$$ Span 1 = \left\{ \begin{matrix} x + 3y \\ 2x + 8y\\ \end{matrix} \right.$$

$$ Span1X=\begin{bmatrix} 1,2 \end{bmatrix}, Span1Y=\begin{bmatrix} 3,8\\ \end{bmatrix}$$




```python
vect1 = np.array([1,2])
vect2 = np.array([3,8])

R = np.arange(-20,20,2)
c1, c2 = np.meshgrid(R,R) 
vectR = vect1 + vect2  
spanRx = c1*vect1[0] + c2*vect2[0] 
spanRy = c1*vect1[1] + c2*vect2[1] 

plt.scatter(spanRx, spanRy, alpha=1,edgecolor='blue', s =5)


plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.grid()
plt.show()
```

## Span 2 (RANK 3)
$$Span 2 = \left\{ \begin{matrix} 4x + 10y + 5z \\ 2x - 5y + 2z\\ \end{matrix} \right.$$


$$Span2X=\begin{bmatrix} 4,2 \end{bmatrix}, Span2Y=\begin{bmatrix} 10,-5\\ \end{bmatrix}, Span2Z=\begin{bmatrix} 5,2\\ \end{bmatrix}$$



```python
vect1 = np.array([4,2,])
vect2 = np.array([10,-5])
vect3 = np.array([5,2])

R = np.arange(-20,20,2)
c1, c2, c3 = np.meshgrid(R,R,R) 
vectR = vect1 + vect2 + vect3
spanRx = c1*vect1[0] + c2*vect2[0] + c3*vect3[0]
spanRy = c1*vect1[1] + c2*vect2[1] + c3*vect3[1]

plt.scatter(spanRx, spanRy, alpha=1,edgecolor='red', s =5)


plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.grid()
plt.show()
```

## Span 3 (RANK 4)
$$Span 3 = \left\{ \begin{matrix} 8x - 5y + 8z - t \\ 10x + 5y - 2z + 3t\\ \end{matrix} \right. $$
 
$$Span3X=\begin{bmatrix} 8,10 \end{bmatrix}, Span3Y=\begin{bmatrix} -5,5\\ \end{bmatrix}, Span3Z=\begin{bmatrix} 8,-2\\ \end{bmatrix}, Span3T=\begin{bmatrix} -1,3\\ \end{bmatrix}  $$

```python
vect1 = np.array([8,10])
vect2 = np.array([-5,5])
vect3 = np.array([8,-2])
vect4 = np.array([-1,3])

R = np.arange(-10,10,2)
c1, c2, c3, c4 = np.meshgrid(R,R,R,R) 
vectR = vect1 + vect2 + vect3 + vect4
spanRx = c1*vect1[0] + c2*vect2[0] + c3*vect3[0] + c4*vect4[0]
spanRy = c1*vect1[1] + c2*vect2[1] + c3*vect3[1] + c4*vect4[1]

plt.scatter(spanRx, spanRy, alpha=1,edgecolor='green', s =5)


plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.grid()
plt.show()
```
